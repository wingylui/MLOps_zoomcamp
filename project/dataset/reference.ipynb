```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
```

```python
df = pd.read_csv("./output/completed_dataset.csv")
```

```python
df["Date_Sold"] = pd.to_datetime(df["Date_Sold"])
df["Year_Sold"] = df["Date_Sold"].dt.year
df["Month_Sold"] = df["Date_Sold"].dt.month
```

```python
target = "Price"
numeric_features = ['Property_Type', 'Bedrooms', 'Bathrooms', 'Parking_Spaces', 'Land_Size', 'Primary_School_Distance', 
                        'Secondary_School_Distance', 'Distance_to_CBD', 'Distance_to_Coast', 'Secondary_ICSEA', 
                        'Primary_ICSEA', 'Year_Sold', 'Month_Sold']
cat_features = []
```

```python
train_data = df[:30000]
val_data = df[30000:]
```

```python
model = LinearRegression()
```

```python
model.fit(train_data[numeric_features + cat_features], train_data[target])
```

```python
train_preds = model.predict(train_data[numeric_features + cat_features])
train_data['prediction'] = train_preds
```

```python
val_preds = model.predict(val_data[numeric_features + cat_features])
val_data['prediction'] = val_preds
```

```python
print(mean_absolute_error(train_data.Price, train_data.prediction))
print(mean_absolute_error(val_data.Price, val_data.prediction))
```

```python
val_data.to_parquet('reference.parquet')
```
